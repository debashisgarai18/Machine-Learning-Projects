### STEP 1: Data Understanding

```
# importing the libraries and reading the .csv file

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
df = pd.read_csv('/content/train.csv')
df
```

```
# to determine the shape

df.shape
```

### STEP 2: Data Preparation

```
# to fill the missing values

df = df.fillna(0)
```

```
print(df.info())
```

```
# to print the columns

df.columns
```

```
order = list(np.sort(df['label'].unique()))
print(order)
```

```
num_mean = df.groupby('label').mean()
num_mean.head()
```

```
round(df.drop('label',axis = 1).mean(),2)
```

```
df = df.reset_index()
```

```
X = df.iloc[:,2:].values
y = df.iloc[:,0].values
```

```
X.shape
```

```
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```
from sklearn.model_selection import KFold
fold = KFold(n_splits = 5, shuffle = True )
hyper_param = [{'gamma':[1e-2,1e-3,1e-4],
                'C':[1,10,100,1000]}]
```

### STEP 3: Model Building

```
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
svc_rbf = SVC(kernel = 'rbf')
model_cv = GridSearchCV(estimator = svc_rbf, param_grid = hyper_param , scoring = 'accuracy', cv = fold , verbose = 1 , return_train_score= True)
model_cv.fit(X,y)
```

```
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```
y_pred = model_cv.predict(X)
```

```
from sklearn.metrics import accuracy_score
print(accuracy_score(y,y_pred))
```
